## Mesters√©ges adat

```python
def frange(start, stop, lepes):
    lista = []
    x = start
    while x < stop:
        lista.append(x)
        x = x + lepes
    return lista
```

```python
eeg_adat = [0.5*(sin(x)+1) for x in frange(0,15,0.1)]
```
